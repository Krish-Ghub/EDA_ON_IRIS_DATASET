```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset
df = sns.load_dataset("iris")

# ===============================
# Data Exploration
# ===============================
print(df.head())        # Display first 5 rows
print(df.tail())        # Display last 5 rows
print(df.info())        # Info about dataset
print(df.describe())    # Statistical summary
print(df.species.unique())  # Unique species in the dataset
print(df.shape)         # Shape of the dataset

# ===============================
# Handling Missing Values (if any)
# ===============================
# Note: inplace=False means this doesn't change the original df
df["sepal_width"].fillna(df["sepal_width"].median(), inplace=False)
df["sepal_length"].fillna(df["sepal_length"].median(), inplace=False)
df["petal_width"].fillna(df["petal_width"].median(), inplace=False)
df["petal_length"].fillna(df["petal_length"].median(), inplace=False)

# ===============================
# Data Visualization
# ===============================

# Set seaborn style
sns.set_style("darkgrid")

# Line plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 5))
sns.lineplot(x="sepal_length", y="sepal_width", data=df, color="r")
plt.title("Sepal: Length vs Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.tight_layout()
plt.show()

# Line plot: Petal Length vs Petal Width
plt.figure(figsize=(8, 5))
sns.lineplot(x="petal_length", y="petal_width", data=df, color="b")
plt.title("Petal: Length vs Width")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.tight_layout()
plt.show()

# Relational plot: Sepal Width vs Species
sns.relplot(x='sepal_width', y='species', data=df, color="g", height=5, aspect=1.5)
plt.title("Sepal Width vs Species")
plt.xlabel("Sepal Width")
plt.ylabel("Species")
plt.tight_layout()
plt.show()

# Bar plot: Species vs Sepal Length
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='sepal_length', data=df, color="y")
plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.tight_layout()
plt.show()
```

## 📊 Summary of Insights

- The dataset contains 150 rows and 5 columns.
- There are 3 unique species: *setosa*, *versicolor*, and *virginica*.
- No missing values were found in the dataset.
- The average sepal length is highest for *virginica*.
- Sepal width decreases slightly as sepal length increases.
- Petal dimensions show a stronger linear relationship compared to sepal dimensions.
